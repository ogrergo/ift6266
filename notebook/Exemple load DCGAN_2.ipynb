```python
from keras.models import load_model
generator = load_model('2017-04-04 21:43:16.039136_generator.h5')
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import matplotlib.gridspec as gridspec
import numpy as np

def display_imgs(imgs):
    n = imgs.shape[0]
    assert n <= 100
    
    fig = plt.figure(dpi=150)

    for i in range(n):
        sb_plot = fig.add_subplot(10, 10, i + 1)
        sb_plot.imshow(imgs[i, :, :, :])
        sb_plot.axis('off')
    plt.show()

# imgs = (data[0:100, :, :,:] / 2. + 0.5).transpose(0, 2,3,1)

def display_gen_output():
    imgs = generator.predict_on_batch(np.random.normal(loc=0, scale=1., size=(100, 100)))
    imgs = (imgs / 2. + 0.5).transpose(0, 2,3,1)
    display_imgs(imgs)

display_gen_output()
```
